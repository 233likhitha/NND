```python
import tensorflow as tf
import numpy as np
x = np.array([[1.0, 0.0, 1.0]])
ip_hid = np.array([
    [0.5, 0.4],
    [-0.2, 0.1],
    [0.3, -0.5]
])
op_hid = np.array([
    [0.7],
    [0.2]
])
b_hid = np.array([[0.1, -0.1]])
b_out = np.array([[0.05]])
hid_raw = np.dot(x, ip_hid) + b_hid
hid = tf.nn.relu(hid_raw)
output_raw = np.dot(hid, op_hid) + b_out
output = tf.sigmoid(output_raw)
print("Hidden layer output:", hid.numpy())
print("Final output:", output.numpy())
```
