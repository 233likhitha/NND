```python
import numpy as np
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
def sigmoid_derivative(a):
    return a * (1 - a)
np.random.seed(1)
W = np.random.randn(2, 1)
b = np.zeros((1, 1))
X = np.array([[0,0,1,1],[0,1,0,1]])
Y = np.array([[0,0,0,1]])
learning_rate = 0.1
epochs = 1000
for _ in range(epochs):
    # Forward propagation
    Z = np.dot(W.T, X) + b
    A = sigmoid(Z)
    # Loss gradient
    dZ = A - Y
    dW = np.dot(X, dZ.T) / X.shape[1]
    db = np.sum(dZ) / X.shape[1]
    # Update weights
    W -= learning_rate * dW
    b -= learning_rate * db
print("Predictions:")
print(sigmoid(np.dot(W.T, X) + b))
```
