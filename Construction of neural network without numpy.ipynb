```python
import random
import math
# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
# Derivative of sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)
# Training data (AND gate)
X = [[0,0], [0,1], [1,0], [1,1]]
Y = [0, 0, 0, 1]
# Initialize weights and bias
w1 = random.random()
w2 = random.random()
b = random.random()
learning_rate = 0.1
epochs = 1000
# Training loop
for _ in range(epochs):
    for i in range(len(X)):
        x1, x2 = X[i]
        target = Y[i]
        # Forward pass
        z = (w1 * x1) + (w2 * x2) + b
        output = sigmoid(z)
  # Error
        error = target - output
        # Backpropagation (weight update)
        w1 += learning_rate * error * sigmoid_derivative(output) * x1
        w2 += learning_rate * error * sigmoid_derivative(output) * x2
        b  += learning_rate * error * sigmoid_derivative(output)
# Testing
print("Testing Neural Network:")
for x in X:
    z = (w1 * x[0]) + (w2 * x[1]) + b
    print(x, "->", round(sigmoid(z), 3))
```
