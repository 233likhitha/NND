```python
from google.colab import files

import cv2

import matplotlib.pyplot as plt

# The file was already uploaded in a previous cell, so this call is redundant.

# Removing uploaded = files.upload()

uploaded = files.upload()

# Ensure the uploaded file is available on disk

if uploaded:

  # Get the name of the first uploaded file

  uploaded_filename = list(uploaded.keys())[0]

  with open(uploaded_filename, 'wb') as f:

    f.write(uploaded[uploaded_filename])

img = cv2.imread(uploaded_filename) # Use the actual uploaded filename

# Add a check to ensure the image was loaded successfully

if img is None:

  raise FileNotFoundError(f"Error: Could not load image '{uploaded_filename}'. Please ensure the file is uploaded and accessible.")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)

plt.imshow(img_rgb)

plt.axis("off")

plt.title("Real Image")

plt.subplot(1,2,2)

plt.imshow(gray, cmap="gray")

plt.axis("off")

plt.title("Gray Image")

plt.show()
```
