```python
# Import TensorFlow and Keras modules
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Import matplotlib for visualization
import matplotlib.pyplot as plt

# Load the MNIST dataset (handwritten digit images)
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize pixel values from [0, 255] to [0, 1]
x_train = x_train / 255.0
x_test = x_test / 255.0

# Build the neural network model
model = Sequential([
    # Flatten 28x28 image into a 1D vector
    Flatten(input_shape=(28, 28)),

    # Hidden layer
    Dense(64, activation='relu'),

    # Output layer for digits 0–9
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
model.fit(x_train, y_train, epochs=7, batch_size=64)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_accuracy)

# Generate predictions
predictions = model.predict(x_test)

# Select image number 1
index = 1

# Display the test image
plt.imshow(x_test[index], cmap='gray')
plt.title("Predicted Digit: " + str(predictions[index].argmax()))
plt.axis('off')
plt.show()
```
